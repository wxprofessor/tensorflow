```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#gpus = tf.config.experimental.list_physical_devices('GPU')
#tf.config.experimental.set_memory_growth(gpus[0], True)
```

### Define the Model
A model is a trained neural network.


Below is the simplest neural network composed of a single layer, one node, and a single X value:<br clear="all">

`[keras.layers.Dense(units=1, input_shape=[1])]`

<ul>
    <li> layer: keras.layers.Dense</li>
    <li> node: units = 1</li>
    <li> X value: input_shape=[1]
</ul>

So below the method `keras.Sequential` groups a linear stack of layers into a model. However, in this case we are making only a single layer as defined above. So this single line of code has defined the simplest of neural networks :)


```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

### Compile the Network
In order to compile the network we must specify 2 functions:
    <ol>
    <li> loss </li>
    <li> optimizer</li>
    </ol>
The way machine learning works, is that the model will attempt to "guess" the answer, then that guess 
-- loss function measures the guess against known correct answer
-- optimizer function generates another guess based upon how well the loss function determined accuracy

different optimizer and loss functions for different scenarious


```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

### Source the data
Self explanatory :)

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

### Train the Network
The network learns by guessing and then measuring the accuracy of the guess (or the loss), and using the chosen optimizer to make a new guess, and this will continue through the number of epics (tries, or guesses) specified... the more guesses, the better the neural network learns, the more accurate it will become :)

```python
#
model.fit(xs, ys, epochs=500)
```

# Now that the model has learned, use it to predict :) 
In the case above, it learned the relationship between the X and Y arrays, so now we can choose an X, and let the model predict what the Y will be based upon the learning from the network above :)

```python
print(model.predict([10.0]))
```
